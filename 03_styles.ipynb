```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi) 
sin = np.sin(x) 
```

```python
print(plt.style.available)
```

```python

for style in plt.style.available:
    print(style)
    plt.figure()
    plt.style.use(style)
    plt.plot(x, sin, c='b')
```
