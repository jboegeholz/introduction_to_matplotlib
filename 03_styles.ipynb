```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.max_open_warning': 0})  # suppress warning of too many opened plots 
x = np.linspace(0, 2*np.pi) 
sin = np.sin(x) 
```

```python
print(plt.style.available)
```

```python

for style in plt.style.available:
    plt.figure()
    plt.style.use(style)
    plt.xlabel(style)
    plt.plot(x, sin, c='b')
```
